```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
HAM_ID = 1

races_df = pd.read_csv('./f1/races.csv', index_col=0)
results_df = pd.read_csv('./f1/results.csv', index_col=0)
races_df = races_df[(races_df['year'] > 2008) & (races_df['year'] < 2022)]
results_df.head()


```

```python
races_df.head()
```

```python
l = []

for i, j in races_df.iterrows():
  year = j['year']
  date = j['date']
  result = results_df[
    (results_df['raceId'] == i) &
    (results_df['driverId'] == HAM_ID)
  ]
  points = 0
  if not result.empty:
    points = result['points']
    points = points.values[0]

  l2 = [year, date, points]
  l.append(l2)
```

```python
ham_df = pd.DataFrame(l, columns=['year', 'date', 'points'])
# pd.set_option('display.max_rows', None)

ham_df = ham_df[(ham_df['year'] > 2014) & (ham_df['year'] < 2020)]

# ham_df.to_csv('ham-2009-2022.csv')
```

```python
s = {}

for i, j in ham_df.iterrows():
  year = j['year']
  points = j['points']
  if not year in s:
    s[year] = [year, points]
  else:
    old_year = s[year]
    old_year.append(points)
    s[year] = old_year

final_df = pd.DataFrame(s.values()).fillna(0)

final_df
```
